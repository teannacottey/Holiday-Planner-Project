# Holiday Planner Recommendation App

## Project Overview

The Holiday Planner is a Python-based travel recommendation app designed to help users choose from 12 tropical and adventurous holiday destinations. The app compares destinations using weather, nightlife and flight data so users can make a more informed decision based on their preferred travel style.

The project was built using Python, Jupyter/Google Colab notebooks, APIs, web-scraping techniques and a Streamlit user interface.

## Problem Statement

Choosing a holiday destination can be difficult because travellers often need to compare several factors at once, including:

- Weather conditions
- Nightlife and entertainment options
- Flight cost
- Flight duration
- Overall suitability for different traveller types

This project solves that problem by combining multiple datasets into one interactive recommendation app.

## Project Aim

The aim of the project is to recommend suitable holiday destinations by analysing weather, nightlife and flight data for 12 selected locations.

The app supports different traveller profiles, such as users who prioritise budget, nightlife, warm weather or overall balance.

## Data Sources

The project uses three main data categories:

| Data Area | Purpose | Example Fields |
|---|---|---|
| Weather | Compare climate suitability | Temperature, humidity, wind speed, cloudiness |
| Nightlife / Places | Compare restaurants, bars and clubs | Ratings, reviews, location data |
| Flights | Compare cost and travel time | Average price, average duration |
