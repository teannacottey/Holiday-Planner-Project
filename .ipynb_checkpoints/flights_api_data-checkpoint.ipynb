# Flights API Data

This notebook follows the process of transforming collected flight data for each destination. Data was gathered from Apify's Google Flights 
API. Particular focus will be placed on data transformation, converting the data into a clean, unified, and standardised format. 

## Import Python Libraries

```python
#data handling 
import pandas as pd 
import numpy as np 

#file and path management 
from pathlib import Path as path 
```

## Load Dataset

```python
flight_data = pd.read_csv('data/raw_data/flights/destinations_flights_agg_chart.csv')
```

## EDA

```python
flight_data.info()
```

## Transform Data

1. Drop unwanted columns
2. Export flight data - destination x metrics (recommendation) & destination x month x metrics (chart)

### Export Flight Recommendation Data

#### 1. Drop unwanted columns

```python
#drop unwanted cols 
destinations_flights_agg = flight_data[['destination','average_price_gbp','average_duration_hours']]
```

#### 2. Aggregate and Export flight data

```python
destinations_flights_agg = destinations_flights_agg.groupby('destination').agg('mean').reset_index().round(2)
destinations_flights_agg
```

```python
destinations_flights_agg.to_csv('destinations_flights_agg.csv', index=False)
```

### Export Flight Chart Data

#### 1. Drop unwanted columns

```python
#drop unwanted cols 
destinations_flights_agg_chart = flight_data[['destination','month','average_price_gbp','average_duration_hours']]
```

#### 2. Export flight data

```python
destinations_flights_agg_chart.to_csv('destinations_flights_agg_chart.csv', index=False)
```
